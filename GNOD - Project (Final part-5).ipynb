```python
#Authentication to spotipy
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
secrets_file = open("secrets.txt","r")
string = secrets_file.read()

secrets_dict={}
for line in string.split('\n'):
    if len(line) > 0:
        #print(line.split(':'))
        secrets_dict[line.split(':')[0]]=line.split(':')[1].strip()
        
#authentication with secrets
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials

#Initialize SpotiPy with user credentials
sp = spotipy.Spotify(auth_manager=SpotifyClientCredentials(client_id=secrets_dict['clientid'],
                                                           client_secret=secrets_dict['clientsecret']))
```

```python
#Taking only numericals
import numpy as np
numerical_ft = audio_features_df.select_dtypes(np.number).reset_index(drop=True)
```

```python
#Scaling the features dataframe
from sklearn.preprocessing import StandardScaler
#Defining scaler
scaler = StandardScaler()
```

```python
#Fitting numerical features in scaler
X_scaled = scaler.fit_transform(numerical_ft)

#Saving the scaler to use it later in separated jupyter notebook
from pickle import dump
dump(scaler, open('scaler.pkl', 'wb'))


#Displaying scaled features in a data frame
pd.DataFrame(X_scaled,columns=numerical_ft.columns)
```

```python
# #Scaler.fit on numerical features
# X_scaled = scaler.fit_transform(numerical_ft)

# #Saving the scaler to use it later in separated jupyter notebook
# from joblib import dump, load
# dump(scaler, 'std_scaler.bin', compress=True)


# # Now, all features will have the same weight.
# pd.DataFrame(X_scaled,columns=numerical_ft.columns)
```

```python
#Here we create the clusters
from sklearn.cluster import KMeans

#Defining Kmeans model
kmeans = KMeans(n_clusters=9, random_state=1234)
#Fitting scaled data in model
kmeans.fit(X_scaled)

#Saving the model to use it later in separated jupyter notebook

dump(kmeans, open('kmeans.pkl', 'wb'))

# Predicting / assigning the clusters:
clusters = kmeans.predict(X_scaled)
```

```python
#Creating a dataframe with cluster_nb
cluster_nb_df = pd.DataFrame({'cluster_nb': clusters})
```

```python
#Concatenating 'audio_features_df' with 'cluster_nb_df'
audio_features_clustered = pd.concat([audio_features_df, cluster_nb_df], axis=1)
audio_features_clustered
```

```python
#Merging 'artists_final' data frame with 'audio_features_clustered' data frame
df_artists_ft = pd.merge(left=artists_final,
                        right=audio_features_clustered,
                        how='inner',
                        left_on='song_id',
                        right_on='id')
df_artists_ft.reset_index(drop=True)
df_artists_ft
```

```python
#Checking counts of clusters
df_artists_ft['cluster_nb'].value_counts(dropna=False)
```

```python
#Saving clean df in a new csv
df_artists_ft.to_csv('df_artists_ft.csv', index=False)
```

```python
#Second prototype
def proposed_title(top_100, top_10000):
    input_title = input("Enter a song title: ")
    input_title = input_title.strip()
    #Collecting from Spotify inputted song's overall info
    results = sp.search(q=input_title, limit=1)
    #Checking if it's in top_100 list
    if input_title in top_100['title'].values:
        proposal = top_100.sample()
        print('Try this top 100 song:')
        print(proposal['title'].values[0]+' from '+proposal['artist'].values[0])
    #If not in top_100 list, then searching for a cluster recommendation
    elif results['tracks']['items'] != []:
        #Extracting only those needed ('href', 'id', 'name', 'type', 'uri', 'external_urls.spotify', 'song_id', 'song_name')
        tracks = json_normalize(results["tracks"]["items"])
        artists_for_song = json_normalize(tracks.iloc[0]['artists'])
        artists_for_song['song_id'] = tracks.iloc[0]['id']
        input_song_id = artists_for_song['song_id'][0]
        #Collecting from Spotify inputted song's features
        song_ft = sp.audio_features(input_song_id)
        #Storing them in variable 'song_audio_features'
        song_audio_features = json_normalize(song_ft)
        #Extracting only its numericals
        song_audio_features = song_audio_features.select_dtypes(np.number).reset_index(drop=True)
        #Loading scaler
        scaler = load(open('scaler.pkl', 'rb'))
        #Scaling song
        song_scaled = scaler.transform(song_audio_features)
        #loading model
        kmeans = load(open('kmeans.pkl', 'rb'))
        #Predicting cluster
        cluster = int(kmeans.predict(song_scaled))
        #Proposing song
        proposal = top_10000[top_10000['cluster_nb'] == cluster]                                                                         
        proposal = proposal.sample()
        print('Try this recommendation from a Spotify playlist:')
        print(proposal['song_name'].values[0]+' from '+proposal['name'].values[0])
    #If introduced title is not found in Spotify, printing
    else:
        print("Sorry I can't provide you with a recommendation. Check for typos or enter a song again")         
    return

#Testing

 

proposed_title(hot_songs, df_artists_ft)
```
